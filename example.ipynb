```python
from mst_lib import *
import networkx as nx
import matplotlib.pyplot as plt
from fractions import Fraction
from math import factorial
```

```python
k = 5
```

```python
G = nx.complete_graph(k)

# Construct a simple path in the graph
edges = [(i, i+1) for i in range(k-1)]

T = nx.from_edgelist(edges)

pos = nx.spring_layout(G)
plt.figure(figsize=(5,5))
nx.draw(G, with_labels=True, pos=pos, node_size=300, node_color='lightblue')
nx.draw_networkx_edges(G, pos=pos, edgelist=T.edges(), width=4, edge_color='r')
plt.show()
```

```python
T = nx.from_edgelist(edges)
```

```python
prob = compute_mst_prob(
    graph =G,
    tree=T,
    show_progress=True,
    print_method=True
)


print(f"Probability of the MST being a particular path for k={k}: {prob} \u2248 {float(prob)}")
```

```python
for k in range(2,8):
    G = nx.complete_graph(k)

    edges = [(i, i+1) for i in range(k-1)]

    T = nx.from_edgelist(edges)
    prob = compute_mst_prob(
        graph =G,
        tree=T,
    )
    
    print(f"Probability of the MST being any path for k={k}: {prob * Fraction(factorial(k), 2)} \u2248 {float(prob * Fraction(factorial(k), 2))}")
```
